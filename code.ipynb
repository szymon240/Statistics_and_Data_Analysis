```r
dane <- read.csv(url("https://raw.githubusercontent.com/fivethirtyeight/data/master/births/US_births_2000-2014_SSA.csv"))

set.seed(2014)
wielkosc_proby <- 100
proba_2014 <- sample(dane[dane$year == 2014, "births"], size = wielkosc_proby)

summary(dane$births)
sd(dane$births)
hist(dane$births, main = "Histogram ilości urodzeń na przestrzeni lat 2000-2014", xlab = "Ilość urodzeń", ylab = "Częstość")

najwiecej_urodzen <- dane[dane$births == max(dane$births),]
najwiecej_urodzen

podzial_na_dni <- aggregate(births ~ day_of_week, data = dane, FUN = sum)
barplot(podzial_na_dni$births, names.arg = c("Pon", "Wt", "Śr", "Czw", "Pt", "Sob", "Ndz"), xlab = "Dzień tygodnia", ylab = "Ilość urodzeń", main = "Podział liczby urodzeń na dni tygodnia")


podzial_na_miesiace <- aggregate(births ~ month, data = dane, FUN = sum)
barplot(podzial_na_miesiace$births, names.arg = c("St", "Lut", "Mrz", "Kw", "Maj", "Cz", "Lip", "Sier", "Wrz", "Paź", "Lis", "Gr"), xlab = "Dzień tygodnia", ylab = "Ilość urodzeń", main = "Podział liczby urodzeń na miesiące")

urodzeni_w_2000 <- subset(dane, year == 2000)

summary(urodzeni_w_2000$births)
sd(urodzeni_w_2000$births)
hist(urodzeni_w_2000$births, main = "Histogram ilości urodzeń - 2000", xlab = "Ilość urodzeń", ylab = "Częstość")

summary(proba_2014)
sd(proba_2014)
hist(proba_2014, main = "Histogram ilości urodzeń w 2014 - próba losowa", xlab = "Ilość urodzeń", ylab = "Częstość")

suma_urodzen <- aggregate(births ~ year, data = dane, FUN = sum)
plot(suma_urodzen$year, suma_urodzen$births, type = "l", xlab = "Rok", ylab = "Ilość urodzeń", main = "Trend ilości urodzeń w latach 2000-2014")
```
